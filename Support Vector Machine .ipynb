```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
#Define Our Data Set
x =pd.DataFrame(cancer['data'])

x.columns = cancer['feature_names']
x['target'] = cancer['target']
x.head()
```

```python
cancer.feature_names
```

```python
print(cancer.target_names)
```

```python
cancer.target
```

```python
dfo = x[x.target == 0].head(50)
dfo
```

```python
df1 = x[x.target ==1].head(50)
df1
```

```python
x['class'] = x.target.apply(lambda x : cancer.target_names[x])
x.head
```

```python
# Dividing Data INto Test And train
```

```python
x=x.drop(['target','class'],axis='columns')
y=cancer['target']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size = 0.3,random_state = 2)
from sklearn.svm import SVC
cla = SVC()
cla.fit(X_train,Y_train)
pred = cla.predict(X_test)
pred
```

```python
score = cla.score(X_test,Y_test)
print(score)
print("accuracy",metrics.accuracy_score(Y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = np.array(confusion_matrix(Y_test,pred))
cm
```

Tune parameters

1.Regularization (C)

```python
model_c = SVC(C=1)
model_c.fit(X_train,Y_train)
model_c.score(X_test,Y_test)
```

```python
model_c = SVC(C=10)
model_c.fit(X_train,Y_train)
model_c.score(X_test,Y_test)
```

2. Gamma

```python
model_g = SVC(gamma='scale')
model_g.fit(X_train,Y_train)
model_g.score(X_test,Y_test)
```

3.Kernal

```python
model_linear_kernal =SVC(kernel='linear')
model_linear_kernal.fit(X_train,Y_train)
```

```python
model_linear_kernal.score(X_test,Y_test)
```
